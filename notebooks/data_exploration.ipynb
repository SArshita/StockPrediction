```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/AAPL.csv", index_col="Date", parse_dates=True)

# Plot Closing Price
plt.figure(figsize=(12,6))
plt.plot(df["Close"], label="Close Price", color="blue")
plt.title("Apple Stock Price Trend")
plt.legend()
plt.show()
```
